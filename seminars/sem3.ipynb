```python
import yfinance as yf

stock_name = 'AAPL' # Apple ticker
stocks = yf.Ticker(stock_name)

# get historical market data over 10 years
ts = stocks.history(period="10y")

ts
```

# Train test split

```python
ts_train, ts_test = ts['Close'][:2018], ts['Close'][2018:]
```

# Visualization

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.figure(figsize=(17,10))

plt.title(stock_name)
plt.ylabel('Price ($)')
plt.plot(ts_train, label='train')
plt.plot(ts_test, label='test')
plt.legend()
```
